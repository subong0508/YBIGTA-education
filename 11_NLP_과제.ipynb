```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
cinderella = open('./cinderella.txt').read()
print(cinderella)
```

```python
wc = WordCloud(background_color="white", collocations=False, stopwords=stopwords)
wc.generate(cinderella)
```

```python
print(wc.words_)
```

```python
wc.to_image()
```

```python
wc.to_file('./18기 응용통계학과 정재은 WordCloud 과제.png')
```
